```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, random_split
import numpy as np
from tqdm import tqdm
import os

# ==================== FIXED SIMPLE CONVNEXT ====================

class SimpleConvNeXt(nn.Module):
    """Minimal ConvNeXt for 28x28 images - FIXED"""
    def __init__(self, num_classes=10):
        super().__init__()
        
        # Stem: 28x28 -> 14x14
        self.stem = nn.Sequential(
            nn.Conv2d(1, 64, kernel_size=4, stride=2, padding=1),
            nn.LayerNorm([64, 14, 14])  # Norm after stem
        )
        
        # Block 1: 14x14 -> 14x14
        self.block1 = nn.Sequential(
            nn.Conv2d(64, 64, kernel_size=7, padding=3, groups=64),
            nn.LayerNorm([64, 14, 14]),
            nn.Conv2d(64, 256, kernel_size=1),
            nn.GELU(),
            nn.Conv2d(256, 64, kernel_size=1),
        )
        
        # Block 2: 14x14 -> 7x7 (with downsampling)
        self.downsample2 = nn.Conv2d(64, 128, kernel_size=2, stride=2)
        self.block2 = nn.Sequential(
            nn.Conv2d(128, 128, kernel_size=7, padding=3, groups=128),
            nn.LayerNorm([128, 7, 7]),
            nn.Conv2d(128, 512, kernel_size=1),
            nn.GELU(),
            nn.Conv2d(512, 128, kernel_size=1),
        )
        
        # Block 3: 7x7 -> 3x3 (with downsampling)
        self.downsample3 = nn.Conv2d(128, 256, kernel_size=2, stride=2)
        self.block3 = nn.Sequential(
            nn.Conv2d(256, 256, kernel_size=7, padding=3, groups=256),
            nn.LayerNorm([256, 3, 3]),
            nn.Conv2d(256, 1024, kernel_size=1),
            nn.GELU(),
            nn.Conv2d(1024, 256, kernel_size=1),
        )
        
        # Global pooling and head
        self.norm = nn.LayerNorm(256)
        self.head = nn.Linear(256, num_classes)
    
    def forward(self, x):
        # Stem
        x = self.stem(x)  # [B, 64, 14, 14]
        
        # Block 1 with residual
        x = x + self.block1(x)  # [B, 64, 14, 14]
        
        # Block 2 with downsampling and residual
        identity = self.downsample2(x)  # [B, 128, 7, 7]
        x = self.downsample2(x)  # [B, 128, 7, 7]
        x = identity + self.block2(x)  # [B, 128, 7, 7]
        
        # Block 3 with downsampling and residual
        identity = self.downsample3(x)  # [B, 256, 3, 3]
        x = self.downsample3(x)  # [B, 256, 3, 3]
        x = identity + self.block3(x)  # [B, 256, 3, 3]
        
        # Global average pooling
        x = x.mean([-2, -1])  # [B, 256]
        
        # Head
        x = self.norm(x)
        x = self.head(x)
        
        return x

# ==================== SIMPLE DATASET ====================

class SimpleDataset(Dataset):
    def __init__(self, class_files, class_names):
        self.images = []
        self.labels = []
        
        print("Loading data...")
        for label, file_path in enumerate(class_files):
            if not os.path.exists(file_path):
                print(f"  ⚠️ File not found: {file_path}")
                continue
            print(f"  Loading {class_names[label]}")
            data = np.load(file_path).astype(np.float32)
            data = data[:50000] / 255.0
            self.images.append(data)
            self.labels.append(np.full(len(data), label))
        
        self.images = np.concatenate(self.images)
        self.labels = np.concatenate(self.labels)
        print(f"Total: {len(self.labels)} images")
    
    def __len__(self):
        return len(self.labels)
    
    def __getitem__(self, idx):
        img = self.images[idx].reshape(1, 28, 28)
        return torch.FloatTensor(img), torch.LongTensor([self.labels[idx]])[0]
```

```python
def train():
    # Setup
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using: {device}")
    
    print(classes_g)
    classes = classes_g[:10]  # Use the first 10 classes from the loaded list
    class_names = [clas for clas, _ in classes]
    
    # File paths - UPDATE THIS PATH
    data_dir = dataset_path  # CHANGE THIS
    class_files = [dataset_path + '/' + file for _, file in classes]
    
    # Load dataset
    dataset = SimpleDataset(class_files, class_names)
    
    # Split
    train_len = int(0.8 * len(dataset))
    val_len = len(dataset) - train_len
    train_ds, val_ds = random_split(dataset, [train_len, val_len])
    
    # Data loaders - SIMPLE: num_workers=0
    train_loader = DataLoader(train_ds, batch_size=128, shuffle=True, num_workers=0)
    val_loader = DataLoader(val_ds, batch_size=128, shuffle=False, num_workers=0)
    
    # Model
    model = SimpleConvNeXt(num_classes=10).to(device)
    print(f"Parameters: {sum(p.numel() for p in model.parameters()):,}")
    
    # Training setup
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
    
    # Train
    for epoch in range(10):
        print(f"\nEpoch {epoch+1}/10")
        
        # Train
        model.train()
        train_loss = 0
        train_correct = 0
        train_total = 0
        
        for images, labels in tqdm(train_loader, desc="Training"):
            images, labels = images.to(device), labels.to(device)
            
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
            _, predicted = outputs.max(1)
            train_total += labels.size(0)
            train_correct += predicted.eq(labels).sum().item()
        
        train_acc = 100. * train_correct / train_total
        print(f"Train Loss: {train_loss/len(train_loader):.4f}, Acc: {train_acc:.2f}%")
        
        # Validate
        model.eval()
        val_correct = 0
        val_total = 0
        
        with torch.no_grad():
            for images, labels in tqdm(val_loader, desc="Validation"):
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                _, predicted = outputs.max(1)
                val_total += labels.size(0)
                val_correct += predicted.eq(labels).sum().item()
        
        val_acc = 100. * val_correct / val_total
        print(f"Val Acc: {val_acc:.2f}%")
```

```python
if __name__ == "__main__":
    train()
```
